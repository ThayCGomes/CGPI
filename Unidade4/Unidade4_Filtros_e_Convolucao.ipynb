# **Aplicação de convolução e filtro laplaciano utilizando OpenCV**

## Filtro Laplaciano


**Importação das bibliotecas**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

**Seção 1 - Abrir o arquivo de imagem e exibir da forma original**

```python
img = cv2.imread("/content/paisagem.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.show()
```

**Seção 2 - Conversão da imagem pra escala de cinza**

```python
img_cinza = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_cinza, cmap='gray')
plt.show()
```

**Seção 3 - Aplicação do filtro Laplaciano**

```python
laplacian = cv2.Laplacian(img_cinza, cv2.CV_16S, ksize=3)
laplacian = cv2.convertScaleAbs(laplacian)
plt.imshow(laplacian, cmap='gray')
plt.show()
```

**Seção 4 - Subtração**

```python
sub = cv2.subtract(laplacian, img_cinza)
plt.imshow(sub, cmap='gray')
plt.show()
```

## Convolução

**Seção 5 - Realce de Bordas**

```python
kernel = np.array([[1, 0, -1],
                   [2, 0, -2],
                   [1, 0, -1]])
```

```python
convolucao = cv2.filter2D(img, -1, kernel)
```

```python
plt.imshow(convolucao)
plt.show()
```

**Seção 6 - Realce de bordas**

```python
kernel2 = np.array([[1, 2, 1],
                   [0, 0, 0],
                   [-1, -2, -1]])
```

```python
convolucao = cv2.filter2D(img, -1, kernel2)
```

```python
plt.imshow(convolucao)
plt.show()
```
